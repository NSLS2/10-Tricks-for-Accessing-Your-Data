## Data Acquisition

```python
from bluesky import RunEngine
from bluesky.plans import scan
from bluesky.plan_stubs import sleep
from ophyd.sim import det, motor
from bluesky_tiled_plugins import TiledWriter
from tiled.client import from_uri
```

```python
client = from_uri('https://tiled.nsls2.bnl.gov')['tst/sandbox/demo']
tw = TiledWriter(client, batch_size=1)
RE = RunEngine()
RE.subscribe(tw);
```

```python
def plan():
    yield from sleep(5)  # while I change tabs
    for num_points in [3, 7, 11]:
        yield from scan([det], motor, -5, 5, num_points)
```

```python
RE(plan())
```
